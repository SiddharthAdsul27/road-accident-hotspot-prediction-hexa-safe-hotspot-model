```python
df1["date"] = pd.to_datetime(df1["date"],format="%d/%m/%Y")
```

```python
df1["month"] = df1["date"].dt.month
```

```python
df1["is_weekend"]=df1["day_of_week"].isin([5,6]).astype(int)
```

```python
df1["is_night"] = ((df1["hour_of_day"] < 6) | (df1["hour_of_day"] > 20)).astype(int)
```

```python
cell_day=df1.groupby(['h3_cell','date']).size().reset_index(name='accidents_today')
```

```python
start=cell_day["date"].min()
end=cell_day["date"].max()
```

```python
full_dates=pd.date_range(start,end,freq='D').date
```

```python
cells=cell_day['h3_cell'].unique()
```

```python
full = pd.MultiIndex.from_product([cells, full_dates],names=['h3_cell','date']).to_frame().reset_index(drop=True)
```

```python
full
```

```python
cell_day['date'] = pd.to_datetime(cell_day['date']).dt.date
full['date'] = pd.to_datetime(full['date']).dt.date
```

```python
full = full.merge(cell_day,on=['h3_cell','date'],how='left')
```

```python
window=7
full['accidents_today'] = full['accidents_today'].fillna(0)
full=full.sort_values(['h3_cell','date'])
```

```python
full['future_accidents'] = (
    full
    .groupby('h3_cell')['accidents_today']
    .apply(lambda s: s.shift(-1).rolling(window=window, min_periods=1).sum())
    .reset_index(level=0, drop=True)
)
```

```python
full['label']=(full['future_accidents']>0).astype(int)
full=full.dropna(subset=['future_accidents']).reset_index(drop=True)
```

```python
cell_attrs = df1.groupby('h3_cell').agg({
    'speed_limit': 'first',
    'road_type': 'first',
    'first_road_class': 'first',
    'first_road_number': 'first',
    'junction_detail': 'first',
    'junction_control': 'first',
    'second_road_class': 'first',
    'second_road_number': 'first',
    'pedestrian_crossing': 'first',
    'road_surface_conditions': 'first',
    'weather_conditions': 'first',
    'light_conditions': 'first',
    'carriageway_hazards': 'first',
    'urban_or_rural_area': 'first'
}).reset_index()
```

```python
full = full.merge(cell_attrs, on='h3_cell', how='left')
```

```python
full['day_of_week'] = pd.to_datetime(full['date']).dt.dayofweek
full['month'] = pd.to_datetime(full['date']).dt.month
full['is_weekend'] = full['day_of_week'].isin([5,6]).astype(int)
full['is_night'] = 0
```

```python
full['acc_last_7'] = full.groupby('h3_cell')['accidents_today'].transform(
    lambda s: s.rolling(7, min_periods=1).sum().shift(1).fillna(0)
)
full['acc_last_30'] = full.groupby('h3_cell')['accidents_today'].transform(
    lambda s: s.rolling(30, min_periods=1).sum().shift(1).fillna(0)
)
```
